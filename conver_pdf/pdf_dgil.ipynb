```python
import PyPDF2
texto = []
# Open the PDF file in read-binary mode
with open('prueba_2.pdf', 'rb') as pdf_file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfReader(pdf_file)
        # Get the number of pages in the PDF file
    num_pages = len(pdf_reader.pages)

    # Loop through each page and extract the text
    for page_num in range(num_pages):
        # Get the page object
        page = pdf_reader.pages[page_num]

        # Extract the text from the page
        text = page.extract_text()
        #pegar
        texto.append(text)
```

```python
texto
```

```python
activo=texto[0].split('Principio Activo')[1].split('\n')[0].strip().replace(':','').strip()
empresa=texto[0].split('Empresa')[1].split('\n')[0].strip().strip()
pres=texto[0].split('ón:')[1].split('\n')[0].strip().replace(':','').strip()
print("Principio activo: ", activo)
print("Empresa: ", empresa)
print("Presentación: ",pres)
```
