```python
import pandas as pd
from marketscraper import pricesearch
import time
import random
```

```python
def search_products(product_list, tiendas):
    results = []

    # Loop en cada producto
    for product in product_list:
        # Loop en cada tienda
        for index, row in tiendas.iterrows():
            tienda = row['tienda']
            cadena = row['cadena']
            estado = row['entidad']  
            municipio = row['municipio']  

            if cadena == "Chedraui":
                func = pricesearch.chedraui
            elif cadena == "Bodega Aurrera":
                func = pricesearch.bodega
            else:
                print(f"Tienda desconocida: {cadena}")
                continue

            try:
                df = func(product, tienda, estado, municipio)  
                results.append(df)
            except Exception as e:
                print(f"Error buscando {product} en {tienda}: {e}")

            sleep_duration = random.uniform(5, 10)  
            time.sleep(sleep_duration)

    # Concatenar todos los resultados
    final_df = pd.concat(results, ignore_index=True)
    
    return final_df
```

#### Listado de productos a buscar

```python
#Lectura de productos en excel. A partir de la línea 8
pacic = pd.read_excel("https://github.com/labdatos-se/supermercados/blob/main/prod_pacic.XLSX?raw=true", skiprows=9)
pacic.columns = pacic.columns.str.lower()
#Toda la base en minúsculas
pacic = pacic.apply(lambda x: x.astype(str).str.lower())
#Eliminar espacios en blanco
pacic = pacic.apply(lambda x: x.astype(str).str.strip())
#Cambiar registro 29 a "mojarra tilapia"
pacic["producto"] = pacic["producto"].replace("tilapia, mojarra tilapia", "mojarra tilapia")
pacic["producto"] = pacic["producto"].replace("tortilla", "tortilla de maíz")
#Renombrar columnas
pacic = pacic.rename(columns={"presentación comercial":"pres", "unidad de medida":"uni_med"})
#Añadir agua embotellada, pañales, formulas, refresco, suero, antigripales, antidiarreicos a la df
nuevos=["agua embotellada", "pañales", "fórmulas lácteas", "refresco", "suero", "antigripales", "antidiarreicos"]
for product in nuevos:
    pacic.loc[pacic.index.max() + 1] = [product, None, None, None]
pacic
```

#### Listado de tiendas Chedraui y Bodega Aurrera

```python
#Leer archivos de ubicaciones de tiendas
tiendas=pd.read_csv("https://raw.githubusercontent.com/labdatos-se/economia/main/mapa_gro/lista_tiendas_gro%20.csv")
tiendas
```

```python
products_to_search = ["agua embotellada", "antidiarreicos"]
df=search_products(products_to_search, tiendas)
df
```

```python
#Salvar a csv
df.to_csv("precios.csv", index=False, encoding="latin1")
```
