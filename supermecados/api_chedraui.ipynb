```python
import pandas as pd
import requests
```

```python
#Definir headers para hacer el request
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}
```

```python
url="https://www.chedraui.com.mx/api/catalog_system/pub/products/search?fq=productName:tortilla"
response=requests.get(url,headers=headers)
response
```

```python
#Obtener el json
data=response.json()
```

```python
len(data)
```

```python
#Obtener el total de productos de acuerdo al len del json
type(data)
```

```python
data
```

```python
#Obtener productName de cada producto y agregarlo a una lista
productos=[]
precios=[]
for i in range(len(data)):
    productos.append(data[i]['items'][0]['name'])
    precios.append(data[i]["items"][0]["sellers"][0]["commertialOffer"]["Price"])
```

```python
precios
```

```python
#Hacer un dataframe con los productos y precios
df=pd.DataFrame({'producto':productos,'precio':precios})
```
